```python
import pandas as pd
import inventorize3 as inv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
suppliers = pd.read_excel(r'D:\data\vendor\vendor.xlsx', sheet_name = 'HGH1Good Receive Register')
```

```python
suppliers.head()
```

```python
print(suppliers.DIVISION.unique())
```

```python
# splitting vendors by DIVISION
# grouping by particular dataframe column
grouped = suppliers.groupby(suppliers.DIVISION)
```

```python
suppliers_Mens = grouped.get_group("Mens")
suppliers_Ladies = grouped.get_group("Ladies")
suppliers_Kids = grouped.get_group("Kids")
suppliers_Fmcg = grouped.get_group("Fmcg Home & Personal Care")
suppliers_FmcgFood = grouped.get_group("Fmcg-Food")
suppliers_Staples = grouped.get_group("Staples")
suppliers_Accoessories = grouped.get_group("Accoessories")
suppliers_Raphaaa = grouped.get_group("Raphaaa")
suppliers_LeatherAccoessories = grouped.get_group("Leather Accoessories")
suppliers_HouseHold = grouped.get_group("House Hold Appliences")
suppliers_HomeDecor = grouped.get_group("Home Decor")
suppliers_WinterGarments = grouped.get_group("Winter Garments")
suppliers_Packing = grouped.get_group("Packing")
suppliers_Handcraft = grouped.get_group("Handcraft")
suppliers_BrandedItems = grouped.get_group("Branded Items")
suppliers_Cmo = grouped.get_group("CMO")
```

```python
df = suppliers_Mens
df.head()
```

```python
df['Profit'] = (df['RSP'] * df['Rec Qty'])- df['Net Amt']
```

```python
df = df.loc[:, ['Rec Dt', 'VENDOR', 'DEPARTMENT', 'Std Rate', 'RSP', 'Rec Qty', 'Net Amt', 'Profit']]
df = df.sort_values('Rec Dt', ascending=True)
df.head()
df.tail()
```

```python
df_a = df.groupby(['VENDOR', 'DEPARTMENT', 'Std Rate']).agg(Total_Qty=('Rec Qty', 'sum'),
                            Net_Amt=('Net Amt', 'sum'),
                            Profit=('Profit', 'sum') ).reset_index().rename(columns={'Std Rate': 'Std_Rate'})
df_a['Product'] = df_a['DEPARTMENT'] + '-' + df_a['Std_Rate'].map(str)
df_a
```

```python
df_a = df_a.loc[:, ['VENDOR', 'Product', 'Std_Rate', 'Total_Qty', 'Net_Amt', 'Profit']]
```
